```python
import requests
```

```python
response = requests.get('https://www.cntraveler.com/gallery/best-bars-in-paris')
```

```python
response.status_code
```

```python
response.content
```

```python
from bs4 import BeautifulS
```

```python
html = BeautifulSoup(response.content, 'lxml')
```

```python
restaurants = html.find_all('section', {
    'class': 'gallery-item'
})
print(len(restaurants))
```

```python
from pprint import pprint
```

```python
pprint(restaurants[0].prettify())
```

```python
r = restaurants[0]
```

```python
r.find('source').attrs['srcset']
```

```python
r.find('h2').text
```

```python
r.find('p', {'class': 'price'}).text
```

```python
r.find('div', {'class': 'dek'}).text
```

```python
r.find('a').attrs['href']
```

```python
%%timeit
data = []
for r in restaurants:
    data.append({
        'image_url': r.find('source').attrs['srcset'] if r.find('source') else None,
        'name': r.find('h2').text,
        'price': r.find('p', {'class': 'price'}).text,
        'description': r.find('div', {'class': 'dek'}).text,
        'url': r.find('a').attrs['href']
    })
```

```python
%%timeit
data = [{
    'image_url': r.find('source').attrs['srcset'] if r.find('source') else None,
    'name': r.find('h2').text,
    'price': r.find('p', {'class': 'price'}).text,
    'description': r.find('div', {'class': 'dek'}).text,
    'url': r.find('a').attrs['href']
} for r in restaurants]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```
